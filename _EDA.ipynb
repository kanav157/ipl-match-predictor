```python
import pandas as pd
import matplotlib.pyplot as plt 
%pip install seaborn
import seaborn as sns 

matches = pd.read_csv('data/matches.csv')

matches['winner'].value_counts().plot(kind='bar')

pd.crosstab(matches['toss_decision'], matches['result'])

# Load fresh copy for EDA (before encoding)
matches_raw = pd.read_csv('data/matches.csv')

# Now plot using matches_raw instead of matches
sns.countplot(x='winner', data=matches_raw, 
              order=matches_raw['winner'].value_counts().index,
              palette='Set2')
plt.xticks(rotation=90)
plt.title("Wins per Team")
plt.show()
```
