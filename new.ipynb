```python
import pandas as pd

# ---- Load old data ----
old_matches = pd.read_csv('data/matches.csv')

# ---- Standardize old data columns ----
old_clean = pd.DataFrame()
old_clean['team1']         = old_matches['team1']
old_clean['team2']         = old_matches['team2']
old_clean['toss_winner']   = old_matches['toss_winner']
old_clean['toss_decision'] = old_matches['toss_decision']
old_clean['venue']         = old_matches['venue']
old_clean['winner']        = old_matches['winner']

# ---- Load new data ----
df_2023 = pd.read_csv('data/ipl23.csv')
df_2024 = pd.read_csv('data/ipl24.csv')
df_2025 = pd.read_csv('data/ipl25.csv')

# ---- Standardize 2023 ----
df_2023_clean = pd.DataFrame()
df_2023_clean['team1']         = df_2023['Team1']
df_2023_clean['team2']         = df_2023['Team2']
df_2023_clean['toss_winner']   = df_2023['Toss winning team']
df_2023_clean['toss_decision'] = df_2023['Toss Decision']
df_2023_clean['venue']         = df_2023['Venue']
df_2023_clean['winner']        = df_2023['Winning Team']

# ---- Standardize 2024 ----
df_2024_clean = pd.DataFrame()
df_2024_clean['team1']         = df_2024['Team 1']
df_2024_clean['team2']         = df_2024['Team 2']
df_2024_clean['toss_winner']   = None
df_2024_clean['toss_decision'] = None
df_2024_clean['venue']         = df_2024['Venue']
df_2024_clean['winner']        = df_2024['winner']

# ---- Standardize 2025 ----
df_2025_clean = pd.DataFrame()
df_2025_clean['team1']         = df_2025['Team 1']
df_2025_clean['team2']         = df_2025['Team 2']
df_2025_clean['toss_winner']   = None
df_2025_clean['toss_decision'] = None
df_2025_clean['venue']         = df_2025['Venue']
df_2025_clean['winner']        = df_2025['winner']

# ---- Merge all ----
combined = pd.concat([
    old_clean,
    df_2023_clean,
    df_2024_clean,
    df_2025_clean
], ignore_index=True)

# ---- Fix toss decision values ----
combined['toss_decision'] = combined['toss_decision'].replace({
    'bowl': 'field'
})

# ---- Fix RCB name (most important!) ----
rcb_fix = {
    'Royal Challengers Bangalore'  : 'Royal Challengers Bengaluru',
    'Royals Challengers Bangalore' : 'Royal Challengers Bengaluru',
}
for col in ['team1', 'team2', 'toss_winner', 'winner']:
    combined[col] = combined[col].replace(rcb_fix)

# ---- Fix other old team names ----
team_fix = {
    'Kings XI Punjab'      : 'Punjab Kings',
    'Delhi Daredevils'     : 'Delhi Capitals',
    'Deccan Chargers'      : 'Sunrisers Hyderabad',
}
for col in ['team1', 'team2', 'toss_winner', 'winner']:
    combined[col] = combined[col].replace(team_fix)

# ---- Clean venue (remove extra commas and spaces) ----
combined['venue'] = combined['venue'].str.replace(',.*', '', regex=True).str.strip()

# ---- Remove defunct teams ----
current_teams = [
    'Mumbai Indians',
    'Chennai Super Kings',
    'Kolkata Knight Riders',
    'Royal Challengers Bengaluru',
    'Sunrisers Hyderabad',
    'Delhi Capitals',
    'Rajasthan Royals',
    'Punjab Kings',
    'Gujarat Titans',
    'Lucknow Super Giants'
]
combined = combined[
    combined['team1'].isin(current_teams) &
    combined['team2'].isin(current_teams)
]

# ---- Verify ----
print(f"Total matches: {len(combined)}")
print(f"\nAll teams:")
print(sorted(combined['team1'].unique()))
print(f"\nMissing values:")
print(combined.isnull().sum())
print(f"\nSample:")
print(combined.head(3))

# ---- Save ----
combined.to_csv('data/new_matches.csv', index=False)
print("\nnew_matches.csv saved successfully!")
```
